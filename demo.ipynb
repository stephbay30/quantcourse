```python
    def get_st_date(self,code):
        st_df = self.pro.namechange(ts_code=code, fields='ts_code,name,start_date,end_date,change_reason')
        st_df['start_date'] = pd.to_datetime(st_df['start_date'], format='%Y%m%d')
        st_df['end_date'] = pd.to_datetime(st_df['end_date'], format='%Y%m%d', errors='coerce')
        st_df = st_df.sort_values(by='start_date')

        # 补全结束日期
        for i in range(len(st_df) - 1):
            if pd.isna(st_df.iloc[i]['end_date']):
                st_df.iloc[i, st_df.columns.get_loc('end_date')] = st_df.iloc[i + 1]['start_date']

        # 筛选包含"ST"的行
        st_df = st_df[st_df['name'].str.contains('ST')]

        # 初始化时间列表
        time_periods = []

        # 遍历包含ST的行，收集时间段
        for index, row in st_df.iterrows():
            start_date = row['start_date']
            end_date = row['end_date'] if not pd.isna(row['end_date']) else pd.to_datetime('now')
            time_periods.append((start_date, end_date))
        return time_periods

    def get_df_st(self,df):
        flag=True
        df.sort_values(['ts_code', 'trade_date'], inplace=True)
        df.reset_index(drop=True, inplace=True)
        for code, group in tqdm(df.groupby('ts_code')):
            time_periods = self.get_st_date(code)
            # 转换 trade_date 列为 datetime 类型
            group.loc[:, 'trade_date'] = pd.to_datetime(group['trade_date'], format='%Y%m%d')

            # 初始化 is_st 列
            group.loc[:, 'is_st'] = 0  # 使用 .loc 确保操作的是原始数据

            # 检查每个交易日是否在任何ST时间段内
            for start, end in time_periods:
                idx = group[(group['trade_date'] >= start) & (group['trade_date'] <= end)].index
                # 直接在原始 df 上修改
                df.loc[idx, 'is_st'] = 1
                if len(idx) != 0 :
                    flag = False
                if not flag:
                    break
            time.sleep(0.13)

        return df
```
